```python
import pandas as pd
import numpy as np
```

```python
import json
from pprint import pprint

with open('suicide_squad.json') as f:
    data = json.load(f)

#pprint()
suicide_members = data['results']['characters']
```

```python
head_size = 30
```

```python
ss_pd = pd.DataFrame.from_dict(suicide_members)
ss_pd = ss_pd.set_index('id')
ss_pd['team']='Suicide Squad'
ss_pd= ss_pd.head(head_size)
```

```python
with open('injustice_league.json') as f:
    data = json.load(f)

#pprint()
injustice_league_members = data['results']['characters']
```

```python
il_pd = pd.DataFrame.from_dict(injustice_league_members)
il_pd = il_pd.set_index('id')
il_pd['team']='Injustice League'
il_pd= il_pd.head(head_size)
```

```python
with open('x_men.json') as f:
    data = json.load(f)

#pprint()
x_men_members = data['results']['characters']
```

```python
chars['origin'] = chars['origin'].apply(json.dumps)
chars['image'] =  chars['image'].apply(json.dumps)
```

```python
chars
```

```python
origin = list(chars['origin'].apply(json.loads))
image = list(chars['image'].apply(json.loads))

origin = [d['name'] if d else 'Undefined' for d in origin ]
image = [d['icon_url'] if d else 'Undefined' for d in image ]
```

```python
chars['origin'] = origin
chars['image'] = image
# origin, image, deck, site_detail_url, aliases, gender, birth, real_name
chars=chars[['name', 'team', 'origin', 'image', 'deck', 'site_detail_url', 'aliases', 'gender', 'birth', 'real_name']]
chars['aliases'] = chars.apply(lambda row: '; '.join(str(row['aliases']).split('\r\n')[:5]), axis=1)
chars['real_name'] = chars.apply(lambda row: '; '.join(str(row['real_name']).split('\r\n')[:5]), axis=1)
chars.head()
```

```python
#relationships.sort_index(inplace=True)

from sklearn.utils import shuffle
chars = shuffle(chars)
```
