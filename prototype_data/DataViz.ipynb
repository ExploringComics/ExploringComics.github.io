```python
import pandas as pd
import numpy as np
```

```python
import json
from pprint import pprint

with open('suicide_squad.json') as f:
    data = json.load(f)

#pprint()
suicide_members = data['results']['characters']
```

```python
ss_pd = pd.DataFrame.from_dict(suicide_members)
ss_pd = ss_pd.set_index('id')
ss_pd['team']='Suicide Squad'
ss_pd= ss_pd.head(10)
```

```python
with open('injustice_league.json') as f:
    data = json.load(f)

#pprint()
injustice_league_members = data['results']['characters']
```

```python
il_pd = pd.DataFrame.from_dict(injustice_league_members)
il_pd = il_pd.set_index('id')
il_pd['team']='Injustice League'
il_pd= il_pd.head(10)
```

```python
with open('x_men.json') as f:
    data = json.load(f)

#pprint()
x_men_members = data['results']['characters']
```

```python
xm_pd = pd.DataFrame.from_dict(x_men_members)
xm_pd = xm_pd.set_index('id')
xm_pd['team']='X Men'
xm_pd = xm_pd.head(10)
```

```python
with open('characters.json') as f:
    data = json.load(f)
    
#pprint(data)
```

```python
characters = pd.DataFrame.from_dict(data)
characters = characters.set_index('id')
characters=characters['origin']
characters.head()
```

```python
SS = ss_pd.join(characters, how='inner')
```

```python
IL = il_pd.join(characters, how='inner')
```

```python
XM = xm_pd.join(characters, how='inner')
```

```python
chars = pd.concat([SS, IL, XM])
```

```python
chars.head()
```

```python
id_ss=ss_pd.reset_index()['id']
```

```python
from itertools import product
relationships = pd.DataFrame(list(product(id_ss, id_ss)), columns=['id1', 'id2'])
```

```python
id_il=il_pd.reset_index()['id']
```

```python
relationships = pd.concat([relationships, pd.DataFrame(list(product(id_il, id_il)), columns=['id1', 'id2'])])
```

```python
id_xm=xm_pd.reset_index()['id']
relationships = pd.concat([relationships, pd.DataFrame(list(product(id_xm, id_xm)), columns=['id1', 'id2'])])
```

```python
relationships.head()
```

```python
with open('origins.json') as f:
    data = json.load(f)
```

```python
origins = pd.DataFrame.from_dict(data)
origins = origins.set_index('id')
origins.head()
```

```python
chars['origin'] = chars['origin'].apply(json.dumps)
```

```python
chars
```

```python
origin = chars['origin'].apply(json.loads)
origin = list(origin)

#origin = list(map(json.loads, origin))
origin = [d['name'] if d is not None else 'Undefined' for d in origin ]
```

```python
chars['origin'] = origin
chars=chars[['name', 'team', 'origin']]
chars.head()
```

```python
chars.to_csv('characters.csv')
```
